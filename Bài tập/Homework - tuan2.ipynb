2. Tính toán và lưu trữ cụm n-gram

```python
tokens = text.split()                   # Tách từ
n = 2                                   # Kích thước của n-gram
ngram_dict = defaultdict(int)           # Tạo một từ điển để lưu trữ cụm n-gram và số lần xuất hiện tương ứng
for i in range(len(tokens) - n + 1):    # Tạo n-gram từ danh sách các từ
    ngram = tuple(tokens[i:i+n])
    ngram_dict[ngram] += 1
for ngram, count in ngram_dict.items(): # In ra các cụm n-gram và số lần xuất hiện tương ứng
    print(' '.join(ngram), ":", count)
```

3. Xây dựng hàm tính xác suất 1 câu dựa vào bigram

```python
def calculate_sentence_probability(sentence, bigram_dict):
    tokens = sentence.split()
    n = 2                                 # Kích thước của n-gram
    probability = 1.0                     # Xác suất khởi đầu bằng 1
    for i in range(len(tokens) - n + 1):
        ngram = tuple(tokens[i:i+n])      # Tạo bigram từng cặp từ liên tiếp trong câu
        if ngram in bigram_dict:          # Nếu bigram có trong từ điển bigram, tính xác suất của bigram
            prob = bigram_dict[ngram] / sum(bigram_dict[prev_word] for prev_word in bigram_dict if prev_word[0] == ngram[0])  # Tính xác suất của bigram
            probability *= prob           # Cập nhật xác suất của câu bằng tích của các xác suất của bigram
    return probability
sentence = "This story centers around Titania and Oberon"               # Ví dụ về câu cần tính xác suất
probability = calculate_sentence_probability(sentence, ngram_dict)      # Tính xác suất của câu
print("Xác suất của câu là:", probability)
```

4. Giả sử cho trước 1 từ, hãy sinh ra 2 từ tiếp theo có xác suất cao nhất. Ví dụ cho từ w1 hãy sinh tiếp 2 từ thành chuỗi w1 w2 w3 có xác suất lớn nhất.

```python
def generate_next_words(word, bigram_dict, num_words=2):
    potential_bigrams = [bigram for bigram in bigram_dict if bigram[0] == word]             # Lọc ra tất cả các bigram có từ đầu tiên là từ cho trước
    sorted_bigrams = sorted(potential_bigrams, key=lambda x: bigram_dict[x], reverse=True)  # Sắp xếp bigram theo xác suất giảm dần
    next_words = [bigram[1] for bigram in sorted_bigrams[:num_words]]                       # Lấy 2 từ tiếp theo có xác suất cao nhất
    return next_words

word = "The"                                                              # Từ cho trước
next_words = generate_next_words(word, ngram_dict, num_words=2)           # Sinh ra 2 từ tiếp theo với xác suất cao nhất
result_string = " ".join([word] + next_words)                             # Nối thành chuỗi
print(f"2 từ tiếp theo bắt đầu từ '{word}' là: {result_string}")
```
