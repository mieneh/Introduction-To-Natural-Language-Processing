```python
import nltk
from nltk.corpus import stopwords
nltk.download('words')
nltk.download('stopwords')
english_vocab = set(w.lower() for w in nltk.corpus.words.words())
stop_vocab = stopwords.words('english')
```

```python
from collections import Counter
def getKMost(k):
  tokens = word_tokenize(text)
  word_counts = Counter(tokens)
  sorted_word_counts = sorted(word_counts.items(), key = lambda x:x[1], reverse = True)
  i = 0
  for word,count in sorted_word_counts:
    if word in english_vocab and word not in stop_vocab:
      i = i + 1
      print(f'{word}: {count}')
      if i == k:
        break
print(getKMost(20))
```

```python
# Tải bộ dữ liệu Brown Corpus
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('brown')
```

```python
# Import Brown Corpus
from nltk.corpus import brown
# Lấy các loại văn bản trong Brown Corpus
categories = brown.categories()
# In ra một số loại văn bản
print("Categories:", categories)
```

```python
for category in categories[:3]:  # In ra các câu văn bản của 3 loại đầu tiên
    sentences = brown.sents(categories=category)
    print('\n', len(sentences))
    print(f"{category} Sentences:")
    for sentence in sentences[:2]:  # In ra 2 câu đầu tiên của mỗi loại
        #print(" ".join(sentence))
        print(sentence)
```
